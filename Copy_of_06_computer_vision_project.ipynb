### 1.1 Dataset Kustom (Opsional)
Di bawah ini disediakan templat dataset kustom yang mungkin ingin Anda gunakan untuk kode Anda. Ini sepenuhnya opsional.

Sebagai alternatif, Anda dapat meninjau materi tentang Augmentasi Data atau membaca tutorial Pytorch https://pytorch.org/tutorials/beginner/basics/data_tutorial.html#creating-a-custom-dataset-for-your-files

```python
#contoh pembuatan dataset kustom kita sendiri, Anda bisa menggunakan ini jika mau/perlu. Sepenuhnya opsional.
import os
import pandas as pd
from torchvision.io import read_image

class CustomImageDataset(Dataset):
    def __init__(self, annotations_file, img_dir, transform=None, target_transform=None):
        self.img_labels = pd.read_csv(annotations_file)
        self.img_dir = img_dir
        self.transform = transform
        self.target_transform = target_transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
        image = read_image(img_path)
        label = self.img_labels.iloc[idx, 1]
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        return image, label
```

```python
#DATA PREPROCESSING AND AUGMENTATION
#Import the necessary libraries for data processing and define the image transformations

import torch
import torchvision
from torchvision import transforms, datasets
from torch.utils.data import DataLoader, random_split
import pandas as pd

# Define transformations for preprocessing and augmentation
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'test': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}

print("Transformations defined successfully.")
```

```python
#Load CSV file to understand the dataset structure before creating a subset for training and testing
df = pd.read_csv('/content/dataset/FINAL_DATASET.csv')
print(df.head())
print(f'Total samples: {len(df)}')
print(f'Columns: {df.columns.tolist()}')
```

```python
'''
Create a CustomImageDataset class that can download images on-the-fly or
load them if they were downloaded.
Then split the DataFrame into training and testing sets
based on the 'dataset_split' column present in the CSV.
'''
import requests
from PIL import Image
from io import BytesIO
from torch.utils.data import Dataset

class DeepfakeURLDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_url = self.dataframe.iloc[idx]['image_url']
        label = self.dataframe.iloc[idx]['label_numeric']

        try:
            response = requests.get(img_url, timeout=10)
            image = Image.open(BytesIO(response.content)).convert('RGB')
        except Exception as e:
            # Fallback to a blank image if URL fails
            image = Image.new('RGB', (224, 224), color=0)

        if self.transform:
            image = self.transform(image)

        return image, label

# Splitting the dataframe into train and test based on the 'dataset_split' column
train_df = df[df['dataset_split'] == 'train'].reset_index(drop=True)
test_df = df[df['dataset_split'] == 'test'].reset_index(drop=True)

# For the sake of speed in this environment, let's use a smaller subset if needed
# train_df = train_df.sample(500, random_state=42).reset_index(drop=True)
# test_df = test_df.sample(100, random_state=42).reset_index(drop=True)

print(f'Training samples: {len(train_df)}')
print(f'Testing samples: {len(test_df)}')
print('Dataset split and custom class defined.')
```
